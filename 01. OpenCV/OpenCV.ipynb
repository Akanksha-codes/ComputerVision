```python
#@ DOWNLOADING LIBRARIES AND DEPENDENCIES: 
import argparse
import imutils
import cv2
from google.colab.patches import cv2_imshow
```

### **IMAGE PROCESSING:**

```python
#@ LOADING THE IMAGE: 
image = cv2.imread("/content/Cat.jpg")
(h, w, d) = image.shape                         # Getting Image Shape. 
print(f"width={w}, height={h}, depth={d}")      # Inspecting Shape.
cv2_imshow(image)
```

```python
#@ INSPECTING PIXELS IN AN IMAGE: 
(B, G, R) = image[100, 50]                      # Initialization. 
print(f"R={R}, G={G}, B={B}")                   # Inspecting Pixels. 
```

```python
#@ RESIZING IMAGE: 
resized = cv2.resize(image, (200, 200))         # Resizing Image. 
cv2_imshow(resized)                             # Inspecting Resized Image. 
```

```python
#@ RESIZING IMAGE: ASPECT RATIO:
r = 300.0 / w                                   # Initializing AR. 
dim = (300, int(h * r))                         # Initialization. 
resized = cv2.resize(image, dim)                # Resizing Image. 
cv2_imshow(resized)                             # Inspecting Image. 
```

```python
#@ RESIZING IMAGE: ASPECT RATIO: 
resized = imutils.resize(image, width=300)        # Resizing Image. 
cv2_imshow(resized)                               # Inspecting Image.       
```

```python
#@ ROTATING IMAGE: 
center = (w // 2, h // 2)                          # Initializing Center of Image. 
M = cv2.getRotationMatrix2D(center, -45, 1.0)      # Initializing Rotation Matrix.  
rotated = cv2.warpAffine(image, M, (w, h))         # Rotating Image. 
cv2_imshow(rotated)                                # Inspecting Image. 
```

```python
#@ ROTATING IMAGE: CLOCKWISE: 
rotated = imutils.rotate(resized, -45)            # Initializing Roatation of Image. 
cv2_imshow(rotated)                               # Inspecting the Image. 
```

```python
#@ ROTATING IMAGE WITHOUT CLIPPING: CLOCKWISE: 
rotated = imutils.rotate_bound(resized, 45)        # Initializing Rotation of Image. 
cv2_imshow(rotated)                                # Inspecting the Image. 
```

```python
#@ BLURRING THE IMAGE: 
blurred = cv2.GaussianBlur(image, (11, 11), 0)      # Initializing Blurring of Image. 
cv2_imshow(blurred)                                 # Inspecting Image. 
```

```python
#@ DRAWING ON AN IMAGE: RECTANGLE: 
output = image.copy()                                           # Making a Copy. 
cv2.rectangle(output, (100, 60), (250, 160), (0, 0, 255), 2)    # Drawing on an Image. 
cv2_imshow(output)                                              # Inspecting Image. 
```

```python
#@ DRAWING ON AN IMAGE: CIRCLE: 
output = image.copy()                               # Initializing Copy of Image. 
cv2.circle(output, (200, 100), 50, (255, 0, 0), -1) # Creating a Circle. 
cv2_imshow(output)                                  # Inspecting Image. 
```

```python
#@ DRAWING ON AN IMAGE: TEXT: 
output = image.copy()                                       # Initializing Copy. 
cv2.putText(output, "My Cat!!", (10, 20), 
            cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)  # Initializing Text on an Image. 
cv2_imshow(output)                                          # Inspecting Image. 
```

### **IMAGE PROCESSING:**

```python
#@ INITIALIZING IMAGE: 
image = cv2.imread("/content/Cat.jpg")              # Reading the Image. 
cv2_imshow(image)                                   # Inspecting the Image. 

#@ CONVERTING IMAGE TO GRAYSCALE: 
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)      # Converting to Grayscale. 
cv2_imshow(gray)                                    # Inspecting the Image. 
```

```python
#@ INITIALIZING EDGE DETECTION: 
edged = cv2.Canny(gray, 30, 150)                    # Initializing Edges in Image. 
cv2_imshow(edged)                                   # Inspecting the Image. 
```

```python
#@ INITIALIZING THRESHOLDING: 
thresh = cv2.threshold(gray, 225, 225, cv2.THRESH_BINARY)[1]
cv2_imshow(thresh)
```

```python
#@ DETECTING AND DRAWING CONTOURS:
cnts = cv2.findContours(thresh.copy(), cv2.RETR_EXTERNAL, 
                        cv2.CHAIN_APPROX_SIMPLE)               # Detecting Contours in Image. 
cnts = imutils.grab_contours(cnts)                             # Initialization. 
output = image.copy()                                          # Initializing Copy of Image. 
cv2.drawContours(output, cnts, -1, (225, 0, 0), 3)             # Drawing Contours. 
cv2_imshow(output)                                             # Inspecting the Image. 
```

```python
#@ INITIALIZING EROSIONS AND DILATIONS: 
mask = thresh.copy()                            # Initializing Copy of Image. 
mask = cv2.erode(mask, None, iterations=5)      # Initializing Erosions. 
cv2_imshow(mask)                                # Inspecting the Image. 
```

```python
#@ INITIALIZING EROSIONS AND DILATIONS: 
mask = thresh.copy()                            # Initializing Copy of Image. 
mask = cv2.dilate(mask, None, iterations=0)     # Initializing Dilations. 
cv2_imshow(mask)                                # Inspecting the Image. 
```

```python
#@ INITIALIZING MASKING AND BITWISE OPERATIONS: 
mask = thresh.copy()                              # Initializing Copy of Image. 
output = cv2.bitwise_and(image, image, mask=mask) # Initializing Bitwise Masking. 
cv2_imshow(output)                                # Inspecting the Image. 
```
