```python
#@ IMPORTING LIBRARIES AND PACKAGES: 
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
#@ INITIALIZING PARAMETERS: 
labels = ["dog", "cat", "panda"]                                    # Initializing Labels. 
np.random.seed(42)                                                  # Initializing Random Seed. 
W = np.random.randn(3, 3072)                                        # Initializing Random Weights. 
b = np.random.randn(3)                                              # Initializing Ranodm Biases. 

#@ LOADING AN IMAGE:
orig = cv2.imread("/content/Bal1.JPG")                              # Reading Image. 
image = cv2.resize(orig, (32, 32)).flatten()                        # Initializing Flatten. 
scores = W.dot(image) + b                                           # Computing Output. 

#@ IMPLEMENTATION OF SCORING FUNCTION: 
for (label, score) in zip(labels, scores):
    print("[INFO] {}: {:.2f}".format(label, score))                 # Inspecting Label and Score. 
cv2.putText(orig, "Label: {}".format(labels[np.argmax(scores)]), 
            (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.9,
            (255, 0, 0), 2)
cv2_imshow(orig)                                                    # Inspecting Image. 
```
